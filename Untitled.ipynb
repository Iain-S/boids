```python
from random import randint
```

```python
lower = [150, 300]
upper = [200,400]

pos = []
for i in range(10):
    pos.append((randint(lower[0], upper[0]), randint(lower[1], upper[1])))
    
```

```python
lower_v = [-15, -30]
upper_v = [20,40]

vel = []
for i in range(10):
    vel.append((randint(lower_v[0], upper_v[0]), randint(lower_v[1], upper_v[1])))
    
vel
```

```python
from matplotlib import pyplot as plt
plt.xlim(0,1000)
plt.ylim(0,700)

x_val = [p[0] for p in pos]
y_val = [p[1] for p in pos]

scat = plt.scatter(x_val, y_val)
```

```python
def update_scatter(frames):
    for i in range(len(pos)):
        pos[i]  =(pos[i][0] +vel[i][0], pos[i][1] +vel[i][1])
        
    scat.set_offsets(pos)
```

```python
import matplotlib.animation as animation
from IPython.display import HTML
anim = animation.FuncAnimation(scat.figure, update_scatter, frames=50, interval=50)

HTML(anim.to_jshtml())
```
